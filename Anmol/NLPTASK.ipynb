```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#loading the dataset and displaying the first 5 rows
df = pd.read_csv(r'C:\Users\HP\Downloads\IMDB Dataset.csv\IMDB Dataset.csv')
df.head()
```

```python
#checking the number of rows and columns in the dataset
#checking for missing values in the dataset
df.info()
df.isnull().sum()
```

```python
#converting all the reviews to lowercase
df['review'] = df['review'].str.lower()
df.head()
```

```python
#removing html tags from the reviews
import re
def remove_html(text):
    clean = re.sub(r'<.*?>', '', text)
    return clean
df['review'] = df['review'].apply(remove_html)
df.head()
```

```python
#removing urls from the reviews
def remove_urls(text):
    clean = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    return clean
df['review'] = df['review'].apply(remove_urls)
df.head()
```

```python
#removing numbers from the reviews
def remove_numbers(text):
    clean = re.sub(r'\d+', '', text)
    return clean
df['review'] = df['review'].apply(remove_numbers)
df.head()
```

```python
#removing punctuation from the reviews
import string
def remove_punctuation(text):
    clean = text.translate(str.maketrans('', '', string.punctuation))
    return clean
df['review'] = df['review'].apply(remove_punctuation)
df.head()
```

```python
#removing extra whitespace from the reviews
def remove_whitespace(text):
    clean = text.strip()
    return clean
df['review'] = df['review'].apply(remove_whitespace)
df.head()
```

```python
#impoting nltk library and downloading the necessary resources for tokenization
import nltk
from nltk.tokenize import word_tokenize
```

```python
#downloading the necessary resources for tokenization
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#tokenizing the reviews - word tokenization
def tokenize_text(text):
    return word_tokenize(text)
df['review'] = df['review'].apply(tokenize_text)
df.head()
```

```python
#importing stop words from nltk and creating a set of stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
#removing negations from the stop words list as this might affect the sentiment analysis
negations = {"not", "no", "never", "nor", "neither", "n't", "cannot", "can't", "don't", "doesn't", "didn't", "won't", "wouldn't", "shouldn't", "isn't", "aren't", "wasn't", "weren't"}
stop_words = stop_words - negations
```

```python
#function to remove stop words from the reviews
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]
df['review'] = df['review'].apply(remove_stopwords)
df.head()
```

```python
#stemming the reviews using Porter Stemmer
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]  
df['review'] = df['review'].apply(stem_tokens)
df.head()
```

```python
#importing WordNetLemmatizer and downloading the necessary resources for lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
#lemmatizing the reviews
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]
df['review'] = df['review'].apply(lemmatize_tokens)
df.head()
```

```python

#joining the tokens back into a single string for each review
df['clean_review'] = df['review'].apply(lambda words: ' '.join(words))
df[['clean_review','sentiment']].head()
```
